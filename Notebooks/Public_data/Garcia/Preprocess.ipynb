```python
import glob
```

```python
meta_ls = []
counts_ls = []
for i in glob.glob('BEGM/*/meta.tsv'):
    ds = i.split('/')[1]
    meta = pd.read_csv(i, sep = '\t', index_col = 0)
    meta.index = [x + '_' + ds for x in meta.index]
    meta['Dataset'] = 'Garcia'
    meta['Timepoint'] = ds.split('_')[0]
    meta['type'] = ds.split('_')[1]
    meta['Species'] = 'Human'
    meta['Cell-type'] = meta['Cluster']
    
    meta_ls.append(meta)
    
    counts = pd.read_csv(i.replace('meta.tsv', 'exprMatrix.tsv'), sep = '\t', index_col = 0).T
    counts.index = [x + '_' + ds for x in counts.index]
    counts_ls.append(counts)
    
```

```python
pd.concat(meta_ls).to_csv('BEGM_meta.tsv', sep = '\t')
```

```python
pd.concat(counts_ls).replace(np.NaN, 0.0).to_csv('BEGM_counts.tsv', sep = '\t')
```
